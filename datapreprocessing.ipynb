```python
# 创建文件
import os
os.makedirs('data', exist_ok=True)
data_file = os.path.join('data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('''NumRoom, HouseType, Price
NA,NA,1245000
2,Floot,2345666
4,Slate,178100
NA,NA,140000''')
print(data_file)
```

```python
# pandas读取文件
import pandas as pd
data = pd.read_csv(data_file)
print(data.shape)
print(data)
```

```python
# data preparation
input, target = data.iloc[:, :2], data.iloc[:, 2]
print(input)
print(target)
# 把不同的取值拆分成不同的列。
input = pd.get_dummies(input, dummy_na=True)
print(input)
```

```python
# 数值的NA一般用mean值来替代
input = input.fillna(input.mean())
print(input)
```

```python
import torch
X = torch.tensor(input.to_numpy(dtype=float))
Y = torch.tensor(target.to_numpy(dtype=float))
print(X)
print("=============================================================")
print(Y)
```

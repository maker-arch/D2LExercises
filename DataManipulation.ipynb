```python
import torch
```

```python
# 1维tensor，指定float
# x = torch.tensor([3,2,3,4], dtype=torch.float32)
x = torch.arange(12, dtype=torch.float32)
x
# inspect the total number of elements
x.numel()
# tensor shape just attr
x.shape
# reshape
X = x.reshape(3,4)
X
# 元素全初始化为0的tensor 也可以torch.zeros((2,3,4))
torch.zeros(2,3,4)
# 元素全初始化为1的tensor
torch.ones((2,3,4))
# creates a tensor with elements drawn from a standard Gaussian (normal) distribution with mean 0 and standard deviation 1.
torch.randn(2,3,4)
# create a tensor with Python list
```

```python
# index -1 start:end
X[-1],X[1:3]
# write elements of a matrix by specifying indices
X[1,2] = 7
X
# 多个elements复制
X[1:2] = 12
X
```

```python
# operation
# exp
torch.exp(x)
# + - * / **
x = torch.tensor([1.0, 2,4,8])
y = torch.tensor([2,2,2,2])
x + y, x - y, x * y, x ** y
# concatenate on axis0 or axis1
X = torch.arange(12, dtype=torch.float32).reshape(3,4)
Y = torch.tensor([[2,0,1,3.0],[1,2,3,4],[4,3,2,1]])
torch.cat((X,Y), axis=0),torch.cat((X,Y),axis=1)
# 逻辑
X == Y,X < Y,X > Y
# sum
# X,X.sum(),X.sum(axis=0),X.sum(axis=1)
```

```python
# broadcasting
# x = torch.arange(3).reshape(3,1)
# y = torch.arange(2).reshape(1,2)


x = torch.arange(12).reshape(3,4,1)
y = torch.arange(8).reshape(1,4,2)

x,y,x + y
```

```python
# saving memory, edit in place
before = id(Y)
Y = Y + X
id(Y) == before

Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = Y + X
print('id(Z):', id(Z))

before = id(X)
X += Y
id(X) == before
```

```python
# Conversion to Other Python Objects
# to numpy
A = X.numpy()
B = torch.from_numpy(A)
B
# to python scalar
#
x = torch.tensor([4.0])
x, x.item(), float(x), int(x)
```
